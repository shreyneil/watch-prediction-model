```python
import pathlib
```

```python
import numpy as np
```

```python
from fastai.vision import (
    ImageList, 
    get_transforms,
    models,
    error_rate,
    imagenet_stats, 
    cnn_learner, 
    ClassificationInterpretation
)
```

```python
chrono_image_path = pathlib.Path('/home/shreyash/Downloads/chrono-analyzer/scraper/images')
```

```python
np.random.seed(42)
```

```python
data = ( 
       ImageList.from_folder(chrono_image_path)
       .split_by_rand_pct()
       .label_from_re(r'.*\/(.+)-.+-.+.jpg')
       .transform(get_transforms())
       .databunch()
       .normalize(imagenet_stats)
)
```

```python
data
```

```python
data.show_batch(rows=3, figsize=(7,7))
```

```python
data.classes
```

```python
learner = cnn_learner(data, models.resnet34, metrics = error_rate)
```

```python
type(learner)
```

```python
learner.fit_one_cycle(5)
```

```python
interpretation = ClassificationInterpretation.from_learner(learner)
```

```python
interpretation.plot_confusion_matrix(figsize = (15,15), dpi = 60)
```

```python
interpretation.most_confused(min_val = 2)
```

```python
interpretation.plot_top_losses(9,figsize=(15,15))
```
